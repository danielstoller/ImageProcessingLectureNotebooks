```python
import os
import matplotlib.pyplot as plt
import cv2
import numpy as np

import mediapipe as mp
from mediapipe.tasks import python
from mediapipe.tasks.python import vision
```

```python
faculty = cv2.imread(os.path.relpath('../../Data/AlanGeorge.jpeg'))
faculty = cv2.cvtColor(faculty,cv2.COLOR_BGR2RGB)

plt.imshow(faculty)
```

```python
base_options = python.BaseOptions(model_asset_path='face_stylizer.task')
options = vision.FaceStylizerOptions(base_options=base_options)

# Create a face stylizer instance with the image mode:
stylizer_ink = vision.FaceStylizer.create_from_options(vision.FaceStylizerOptions(base_options=
                                                                           python.BaseOptions(model_asset_path='face_stylizer_color_ink.task')))

stylizer_sketch = vision.FaceStylizer.create_from_options(vision.FaceStylizerOptions(base_options=
                                                                              python.BaseOptions(model_asset_path='face_stylizer_color_sketch.task')))

stylizer_oil = vision.FaceStylizer.create_from_options(vision.FaceStylizerOptions(base_options=
                                                                           python.BaseOptions(model_asset_path='face_stylizer_oil_painting.task')))

mp_image = mp.Image(image_format=mp.ImageFormat.SRGB, data=faculty)

face_stylizer_ink = stylizer_ink.stylize(mp_image)
face_stylizer_sketch = stylizer_sketch.stylize(mp_image)
face_stylizer_oil = stylizer_oil.stylize(mp_image)
```
